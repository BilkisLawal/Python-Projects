```python
# Analying what continents have grown faster that others
```

```python
df.groupby('Continent')[numeric_df.columns].mean()
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
df2 = df.groupby('Continent')[numeric_df.columns].mean().sort_values(by= '2022 Population', ascending = False)
df2
```

```python
df2.plot()
```

```python
df3 = df2.transpose()
df3
```

```python
df3.plot(figsize = (20,7))
```

```python
numeric_df.columns
```

```python
df4 = df.groupby('Continent')[['1970 Population', '1980 Population', '1990 Population',
       '2000 Population', '2010 Population', '2015 Population',
       '2020 Population', '2022 Population']].mean().sort_values(by= '2022 Population', ascending = False)
df4
```

```python
df41 = df4.transpose()
```

```python
df41.plot(figsize = (20,10))
```
