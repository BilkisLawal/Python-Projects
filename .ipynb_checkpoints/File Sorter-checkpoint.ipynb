# Automated File Sorter in File Explorer

```python
import os, shutil
```

```python
path = r"/Users/Bilkis/Downloads/"
```

```python
os.listdir(path)
```

```python
Names_folders = ['Excel files', 'PDF files', 'Word files', 'Powerpoint files', 'Photos']
```

```python
for loop in range(0,5):
    os.path.exists(path + Names_folders[loop])
```

```python
for loop in range(0,5):
   if not os.path.exists(path + Names_folders[loop]):
       print(path + Names_folders[loop])
       os.makedirs(path + Names_folders[loop])
```

```python
doc_name = os.listdir(path)
```

```python
for file in doc_name:
    if ".csv" and ".xlsx" in file and not os.path.exists(path + "Excel files/" + file):
        shutil.move(path + file, path + "Excel files/" + file)
    elif  ".pdf" in file and not os.path.exists(path + "PDF files/" + file):
        shutil.move(path + file, path + "PDF files/" + file)
    elif ".jpg" and ".mp4" in file and not os.path.exists(path + "Photos/" + file):
        shutil.move(path + file, path + "Photos/" + file)
    elif ".png" and ".jpeg" in file and not os.path.exists(path + "Photos/" + file):
        shutil.move(path + file, path + "Photos/" + file)
    elif ".docx"  in file and not os.path.exists(path + "Word files/" + file):
        shutil.move(path + file, path + "Word files/" + file)
    elif ".pptx"  in file and not os.path.exists(path + "Powerpoint files/" + file):
        shutil.move(path + file, path + "Powerpoint files/" + file)
    else:
        'file not included'
```
