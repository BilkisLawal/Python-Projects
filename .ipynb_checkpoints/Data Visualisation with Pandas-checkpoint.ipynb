# Data Visualisation with Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'/Users/Bilkis/Downloads/Ice Cream Ratings.csv')
df
```

```python
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('_classic_test_patch')
```

```python
df.plot(figsize = (10,2))
```

```python
df.plot(kind = 'hist')
```

```python
df.plot(kind = 'hist',bins = 20)
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'line', title = 'Icecream Ratings', xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'bar', stacked = True)
```

```python
df['Flavor Rating'].plot(kind = 'bar')
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.scatter(x = 'Flavor Rating', y = 'Overall Rating')
```

```python
df.plot.scatter(x = 'Flavor Rating', y = 'Overall Rating', s = 500, c ='red')
```

```python
df.plot.area()
```

```python
df.plot.area(figsize =(10,5))
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10,6))
```
